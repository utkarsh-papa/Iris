### ML alogo analysis

```python
df.head()
```

```python
# spliting the data into train and test
X=df.iloc[:,0:4]
y=df.iloc[:,4:]
```

```python
X.head()
```

```python
y['Species'] = y['Species'].map( {'Iris-setosa': 0, 'Iris-versicolor': 1,'Iris-virginica':2} ).astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
print('size of train data=',X_train.shape,'\nsize of test data=',X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
y_pred = pd.DataFrame(y_pred, columns =['Species'])
```

```python
knn.score(X_test,y_test)
```

### Score Of KNeighborsClassifier is 95.55%

## Confusion matrix and classification_report

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_test, y_pred))
```

## Cross Validation


```python
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(knn,X,y,cv=5)
```

```python
print(cv_score)
```

### Applying Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
lg.score(X_test,y_test)
```

### Score Of Logistics regression 97.77%

## GridSearchCV in scikit-learn


```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': np.arange(1, 50)}
knn_cv = GridSearchCV(knn, param_grid, cv=5)
knn_cv.fit(X_train, y_train)
knn_cv.best_params_
```

```python
knn_cv.best_score_
```

### Using GridsearchCv score is 98%

## Tree alogorithm

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier(max_depth=2, random_state=1)
dt.fit(X_train,y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
